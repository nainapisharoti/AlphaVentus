# Import Modules

```python
import os
import sys
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
plt.style.use('seaborn-white')
```

# Load NetCDF From Cluster

```python
netcdf_file_loc_will = '/p/lustre2/nainap/'
wrf_domain6_file = os.path.join(netcdf_file_loc_will,'From_Will', 'wrfout_domain6')
```

```python
wrf_domain6 = xr.open_dataset(wrf_domain6_file)
```

# Setting Up Variables

```python
yz_plane_cols, yz_plane_rows = np.meshgrid(wrf_domain6['south_north'], wrf_domain6['bottom_top'])
xz_plane_cols, xz_plane_rows = np.meshgrid(wrf_domain6['west_east_stag'], wrf_domain6['bottom_top'])
xy_plane_cols, xy_plane_rows = np.meshgrid(wrf_domain6['west_east_stag'], wrf_domain6['south_north'])
```

```python
qoi = 'U'
qoi_dim = wrf_domain6[qoi].shape
#shape format in : [time, z, y, x]
[time_dim, bottom_top_dim, south_north_dim, west_east_stag_dim] = qoi_dim 
```

```python
## Set-up different frames for plotting (YZ-plane)
location_spec_flag = 0 # used to provide specific locations

if location_spec_flag == 0:
    qoi_yz_1 = wrf_domain6[qoi].isel(Time=0).isel(west_east_stag=0) 
    qoi_yz_2 = wrf_domain6[qoi].isel(Time=0).isel(west_east_stag=int(west_east_stag_dim*0.5))
    qoi_yz_3 = wrf_domain6[qoi].isel(Time=0).isel(west_east_stag=int(west_east_stag_dim*0.75))
```

```python
## Set-up different frames for plotting (XZ-plane)
location_spec_flag = 0 # used to provide specific locations

if location_spec_flag == 0:
    qoi_xz_1 = wrf_domain6[qoi].isel(Time=0).isel(south_north=0) 
    qoi_xz_2 = wrf_domain6[qoi].isel(Time=0).isel(south_north=int(south_north_dim*0.5))
    qoi_xz_3 = wrf_domain6[qoi].isel(Time=0).isel(south_north=int(south_north_dim*0.75))
```

```python
## Set-up different frames for plotting (XY-plane)
location_spec_flag = 0 # used to provide specific locations

if location_spec_flag == 0:
    qoi_xy_1 = wrf_domain6[qoi].isel(Time=0).isel(bottom_top=0) 
    qoi_xy_2 = wrf_domain6[qoi].isel(Time=0).isel(bottom_top=int(bottom_top_dim*0.5))
    qoi_xy_3 = wrf_domain6[qoi].isel(Time=0).isel(bottom_top=int(bottom_top_dim*0.75))
```

# Contour Plots

```python
cmap_name = 'rainbow'
south_north_label = 'south_north_index'
west_east_label   = 'west_east_index'
bottom_top_label   = 'bottom_top_index'
```

### YZ-Plane

```python
f, axs = plt.subplots(1, 3, figsize = (20, 5) )

plt.subplot(1, 3, 1)
plt.contourf(yz_plane_cols, yz_plane_rows, qoi_yz_1, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(south_north_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': west_east_stag = 0')

plt.subplot(1, 3, 2)
plt.contourf(yz_plane_cols, yz_plane_rows, qoi_yz_2, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(south_north_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': west_east_stag = {}'.format(int(west_east_stag_dim*0.5)))

plt.subplot(1, 3, 3)
plt.contourf(yz_plane_cols, yz_plane_rows, qoi_yz_3, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(south_north_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': west_east_stag = {}'.format(int(west_east_stag_dim*0.75)))
```

### XZ-Plane

```python
f, axs = plt.subplots(1, 3, figsize = (20, 5) )

plt.subplot(1, 3, 1)
plt.contourf(xz_plane_cols, xz_plane_rows, qoi_xz_1, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': south_north = 0')

plt.subplot(1, 3, 2)
plt.contourf(xz_plane_cols, xz_plane_rows, qoi_xz_2, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': south_north = {}'.format(int(south_north_dim*0.5)))

plt.subplot(1, 3, 3)
plt.contourf(xz_plane_cols, xz_plane_rows, qoi_xz_3, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(bottom_top_label)
plt.title(qoi + ': south_north = {}'.format(int(south_north_dim*0.75)))
```

### XY-Plane

```python
f, axs = plt.subplots(1, 3, figsize = (20, 5) )

plt.subplot(1, 3, 1)
plt.contourf(xy_plane_cols, xy_plane_rows, qoi_xy_1, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(south_north_label)
plt.title(qoi + ': bottom_top = 0')

plt.subplot(1, 3, 2)
plt.contourf(xy_plane_cols, xy_plane_rows, qoi_xy_2, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(south_north_label)
plt.title(qoi + ': bottom_top = {}'.format(int(bottom_top_dim*0.5)))

plt.subplot(1, 3, 3)
plt.contourf(xy_plane_cols, xy_plane_rows, qoi_xy_3, 20, cmap=cmap_name);
plt.colorbar();
plt.xlabel(west_east_label)
plt.ylabel(south_north_label)
plt.title(qoi + ': bottom_top = {}'.format(int(bottom_top_dim*0.75)))
```
